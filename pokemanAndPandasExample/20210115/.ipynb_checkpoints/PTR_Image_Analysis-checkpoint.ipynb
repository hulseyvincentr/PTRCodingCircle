```python
from scipy.io import loadmat
from pathlib import Path
import tifffile
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from skimage.measure import label, regionprops, find_contours
from skimage.morphology import closing, square, disk, opening, dilation, erosion, watershed, label
from skimage.color import label2rgb 
from scipy.ndimage import binary_fill_holes
import scipy
```

```python
#get tif and mat files
mat_file = [cmat for cmat in Path.cwd().glob('*.mat')]
tif_file = [cmat for cmat in Path.cwd().glob('*.tif')]
```

## Loading and selectin an ROI

```python
timeseries = tifffile.imread(str(tif_file[0]))
print(timeseries.shape)
```

### Sum intensity image

```python
sum_img =  np.sum(timeseries, axis=0)
plt.imshow(sum_img, clim=[0,2000])
plt.colorbar()
```

### Thresholding image for roi selection  
Then clean up image using morphological filters.

```python
#setting threshold
binary_img = np.zeros(sum_img.shape)
binary_img[sum_img > 1000] = 1
plt.imshow(binary_img)
plt.colorbar()
```

```python
selem = disk(3)
close_image = closing(binary_img, selem)
plt.imshow(close_image)
```

```python
selem = disk(3)
open_image = opening(close_image, selem)
plt.imshow(open_image)
```

```python
label_img = label(open_image)
regions = regionprops(label_img)
```

```python
#generate an index for each ori from flattened image
roi_list = []
for creg in regions:
    #print(creg)
    mask = np.zeros(sum_img.shape)
    mask[creg.coords[:, 0], creg.coords[:, 1]] = 1
    mask = binary_fill_holes(mask)
    #selem = disk(3)
    #mask = erosion(mask, selem)
    #fig = plt.figure()
    #ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
    #ax1.imshow(mask)
    #roi_list.append(np.where(mask.flatten()))

    roi_list.append(np.where(mask.flatten()))
```

```python

roi_list_df = pd.DataFrame(data=[roi_list], columns=range(len(roi_list)), index = ['roi_index']).T
roi_list_df
```

```python
fig = plt.figure(figsize=[5,5])
ax = plt.axes([0.1,0.1,0.9,0.9])
img1 = ax.imshow(sum_img, clim=[0,2000])
fig.colorbar(img1)
for row, dseries in roi_list_df.iterrows():
    maskimage = np.zeros(sum_img.shape)
    index = sum_img.shape[0]*sum_img.shape[1]
    maskimage.reshape(index)[dseries['roi_index'] ] = 1
    contours = find_contours(maskimage, 0.8)
    for n, contour in enumerate(contours):
        contour = np.array(contour, dtype = np.int)
        #ax1.plot(contour[:, 1], contour[:, 0], linewidth = 2, color = colors[i])
        c = ax.plot(contour[:, 1], contour[:, 0]+1, linewidth = 2, color = 'red')
    indexC = scipy.ndimage.center_of_mass(maskimage)
    line = ax.text(indexC[1], indexC[0], str(row), color = 'black')
fig.savefig('Example.png')
#plt.show(graph)
```

### Viewing and selecting ROI

```python
#create a backound roi
x = np.arange(400,500)
y = np.arange(150, 500)
xv, yv = np.meshgrid(x, y)
mask = mask = np.zeros(sum_img.shape)
mask[yv.flatten().astype(np.int), xv.flatten().astype(np.int)] = 1
plt.imshow(mask)
background = pd.Series({'roi_index' : np.where(mask.flatten())}, name='background')
roi_list_df.loc['background'] = background
```

```python
fig = plt.figure(figsize=[5,5])
ax = plt.axes([0.1,0.1,0.9,0.9])
img1 = ax.imshow(sum_img, clim=[0,2000])
fig.colorbar(img1)
for row, dseries in roi_list_df.iterrows():
    maskimage = np.zeros(sum_img.shape)
    index = sum_img.shape[0]*sum_img.shape[1]
    maskimage.reshape(index)[dseries['roi_index'] ] = 1
    contours = find_contours(maskimage, 0.8)
    for n, contour in enumerate(contours):
        contour = np.array(contour, dtype = np.int)
        #ax1.plot(contour[:, 1], contour[:, 0], linewidth = 2, color = colors[i])
        c = ax.plot(contour[:, 1], contour[:, 0]+1, linewidth = 2, color = 'red')
    indexC = scipy.ndimage.center_of_mass(maskimage)
    line = ax.text(indexC[1], indexC[0], str(row), color = 'black')
fig.savefig('Example.png')
```

```python
all_signal = []
for row, dseries in roi_list_df.iterrows():
    signal = np.zeros([timeseries.shape[0]], dtype = np.float)
    for layer in range(timeseries.shape[0]):
        cimage = timeseries[layer].flatten()
        signal[layer] = np.mean(cimage[dseries['roi_index']])
    all_signal.append(signal)
roi_list_df['signal'] = all_signal
signal_Back = []
#remove background
for row,dseries in roi_list_df.iterrows():
    signal_Back.append(dseries['signal'].astype(np.float) - roi_list_df['signal'].loc['background'].astype(np.float))
roi_list_df['signal-background'] = signal_Back
```

```python
roi_list_df
for row,dseries in roi_list_df.iterrows():
    print(dseries['roi_index'][0].shape)
    if dseries['roi_index'][0].shape[0] < 3000:
        roi_list_df.drop(row, inplace=True)
        print('Dropping', row)
```
