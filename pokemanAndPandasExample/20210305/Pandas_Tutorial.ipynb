# 3. Processing the data
1. What is the best pokemon? Sum all the points to give a maximum score.
2. Are legendary better the others?

```python
#1 finding the ultimate pokeman
#not the fastest method but the easiest to understand

score = {}
for row, dseries in poke_csv.iterrows():
    score[row] = dseries[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum()
```

```python
#making a dataframe out of a dictionary
score2 = pd.DataFrame.from_dict(score, orient='index', columns=['score'])
score2
```

```python
# add the score a s new column to poke_csv
poke_csv2 = pd.concat([poke_csv, score2], axis=1)
poke_csv2
```

```python
poke_csv2.sort_values(['score'], ascending=False, inplace=True)
poke_csv2
```

```python
# 4 Outputing the data
```

```python

poke_csv2.to_excel('SortedByBestScore.xlsx')
```

```python
plt.figure(figsize=(40,5))
poke_csv2['score'][:30].plot.bar()
plt.savefig('BestPOKE.svg')
```

# 3 Processing: Is legendary the best type?

```python
legend = {}
for cgroup, cframe in poke_csv2.groupby(['Legendary']):
    print(cgroup)
    legend[cgroup] = cframe.mean()
```

```python
legend2 = pd.DataFrame.from_dict(data=legend)
```

```python
legend2
```

```python
legend2.rename({True : 'Legendary', False: "Other"}, inplace=True, axis=1)
legend2.drop(['Legendary', 'Generation'], axis=0, inplace=True)
legend2
```

```python
legend2.plot.bar()
plt.savefig('Legendary.png')
```
